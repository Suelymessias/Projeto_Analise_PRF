## Objetivo da Análise

Este projeto tem como objetivo analisar as causas dos acidentes nas rodovias do Brasil e se há relação com o posicionamento dos radares.

- Analisar a eficácia dos radares e o impacto dos radares na redução de acidentes.

- Identificar rodovias com maior concentração de acidentes e avaliar a presença de radares nesses locais.

- Avaliar a gravidade dos acidentes ocorridos em diferentes trechos das rodovias.

- Conduzir uma análise estatística para determinar a correlação entre a presença de radares e a diminuição na frequência de acidentes.

- Analisar os tipos de acidentes que ocirreram nas rodovias ao longo dos anos abordados.

- Conduzir uma analise temporal nos anos abordados.

- Considerar não apenas os aspectos de segurança, mas também as implicações econômicas decorrentes dos acidentes.

## Conclusão

- A alta concentração de acidentes nas proximidades dos radares sugere uma possível ineficácia desses dispositivos ou a falta de sinalização adequada, o que demanda uma revisão na eficiência dos mesmos.

- Registrou-se um aumento significativo nos acidentes ocorridos a uma distância de 6 a 7 km dos radares, especialmente com o tipo de acidente de colisão traseira. Esses números sugerem possíveis problemas de atenção por parte dos condutores ou deficiências na sinalização viária nesses trechos.

- A maioria dos acidentes registrados é do tipo colisão traseira, o que sugere possíveis deficiências na manutenção das estradas e falta de atenção por parte dos condutores.

- Não foi observada uma diminuição significativa nos números de vítimas ao comparar os anos de 2021, 2022 e a primeira metade de 2023. Isso evidencia uma estabilidade preocupante nos índices de acidentes rodoviários com vítimas, demandando atenção contínua às medidas de segurança viária.


## Propostas para Reduzir a Incidência de Acidentes nas Rodovias

- Instalação Estratégica de Radares e Manutenção das Rodovias: Utilizando as informações sobre a distribuição dos acidentes em relação à distância dos radares, as autoridades responsáveis podem considerar a realocação estratégica dos dispositivos de monitoramento, ao mesmo tempo em que investem na manutenção adequada das rodovias para garantir condições seguras de tráfego.

- Avaliação de Locais Críticos: Identificar e avaliar cuidadosamente os locais críticos onde ocorrem picos de acidentes em distâncias maiores dos radares, como os intervalos de 3 a 5 km e 6 a 7 km, torna-se fundamental. Além disso, é necessário analisar a qualidade da infraestrutura viária nesses locais e implementar medidas de manutenção e segurança conforme necessário.

- Melhoria da Sinalização e Conscientização: Investir em medidas adicionais de segurança, como melhorias na sinalização e dispositivos de alerta, é essencial para prevenir acidentes. Paralelamente, campanhas de conscientização e educação para os motoristas são importantes para aumentar a percepção dos riscos e promover comportamentos mais seguros nas estradas, complementando os esforços de manutenção e segurança.

- Monitoramento Contínuo e Análise de Dados: Estabelecer um sistema de monitoramento contínuo e análise de dados é fundamental para acompanhar a eficácia das medidas implementadas. Além de coletar dados sobre acidentes, é importante avaliar regularmente a condição das rodovias e a eficiência dos dispositivos de segurança, permitindo ajustes e adaptações conforme necessário para otimizar a segurança viária.

```python
#Gerando metadados
def generate_metadata(dataframe):
    """
    Gera um dataframe contendo metadados das colunas do dataframe fornecido.

    :param dataframe: DataFrame para o qual os metadados serão gerados.
    :return: DataFrame contendo metadados.
    """

    # Coleta de metadados básicos
    metadata = pd.DataFrame({
        'nome_variavel': dataframe.columns,
        'tipo': dataframe.dtypes,
        'qt_nulos': dataframe.isnull().sum(),
        'percent_nulos': round((dataframe.isnull().sum() / len(dataframe))* 100,2),
        'cardinalidade': dataframe.nunique(),
    })
    metadata=metadata.sort_values(by='tipo')
    metadata = metadata.reset_index(drop=True)

    return metadata

metadata_df = generate_metadata(df_concat)
metadata_df
```

Nesta etapa será feita a limpeza dos dados, implementando as transformações essenciais identificadas durante a visualização, como tratamento de valores ausentes, ajustes no formato das coordenadas de latitude e longitude, além de otimização do formato das datas.

### Dados Acidentes nas Rodovias PRF


Durante esta etapa, serão implementados os devidos tratamentos, abordando:

**Tratamento de valores nulos:**
- Em variáveis do tipo "object" a estratégia adotada consistiu em preencher os valores ausentes com "desconhecido" com o intuito de preservar a integridade dos dados.

- Para variáveis `br` substitui esses espaços vazios por "000". Tal escolha foi deliberada, uma vez que esse valor destaca-se como discrepante em relação aos demais dados das colunas, proporcionando facilidade na identificação durante as análises subsequentes.


**Ajustes e Padronizações nas Colunas:**

- Em relação às variáveis de `latitude` e `longitude`, procedeu-se à substituição da vírgula por ponto, promovendo a adequação do tipo das variáveis para permitir análises mais precisas no futuro.

- Para as colunas `data_inversa` e `horario`, efetuou-se a conversão para o formato datetime, proporcionando maior facilidade no manuseio das informações de data e hora.

- A variável `km` passou por uma divisão, resultando em duas novas colunas: `Km`, destinada a representar os valores em quilômetros, e `m`, destinada a representar os valores em metros relativos à localização do acidente. Transformei a mesma em float e consequentemente os valores nulos foram preenchidos por "000".

- A coluna `br` será renomeada para 'rodovia' para alinhar-se com a coluna correspondente na base de dados dos radares.



#### Tratamento de Valores Nulos

```python
# Divisão da coluna Km e alteração para o formato float
df_prf01[['Km', 'm']] = df_prf01['km'].str.split(',', expand=True).astype(float)
```

```python
# Preenchimento de valores nulos das variáveis object
var_categoricas = df_prf01.select_dtypes(include=['object']).columns
df_prf01[var_categoricas] = df_prf01[var_categoricas].fillna('desconhecido')
```

```python
# Preenchimento de valores nulos das variáveis numéricas
var_numericas = df_prf01.select_dtypes(include=['float64','int64']).columns
df_prf01[var_numericas] = df_prf01[var_numericas].fillna(000)
```

####Adequação e Padronização nas Colunas

```python
# Ajustando valores das colunas latitude e longitude. Alterando tipo para numérico.
df_prf01['latitude'] = df_prf01['latitude'].str.replace(',','.').astype(float)
df_prf01['longitude'] = df_prf01['longitude'].str.replace(',','.').astype(float)
```

```python
# Alterando as colunas horário e data_inversa para datetime.
df_prf01['data_inversa'] = pd.to_datetime(df_prf01['data_inversa'], format='%Y-%m-%d')
df_prf01['horario'] = pd.to_datetime(df_prf01['horario'], format='%H:%M:%S')
```

```python
# Alterando o nome da coluna br para rodovia.
df_prf01= df_prf01.rename (columns={'br': 'rodovia'})
```

```python
# Verificando metadados para validar que as transformações foram efetuadas.
generate_metadata(df_prf01)
```

## Analise da Proximidade dos Radares
- Utilizando a função GeoDataFrame() da biblioteca GeoPandas que permite o tratamento de dados geoespaciais, como latitude e longitude, e a realização de operações espaciais.

- Para calcular a distância entre cada acidente e todos os radares, utilizei a função distance()  e em seguida a função min() para encontrar a distância mínima.

- Apresentação de uma descrição estatística da coluna 'distancia_radar_km', fornecendo informações como média, mediana, mínimo, máximo e quartis das distâncias calculadas.

- Condução de uma análise da distância entre os locais de acidentes e os radares mais próximos. Utilizando histograma, será examinada a distribuição da ocorrência de acidentes em relação à proximidade dos radares. Essa análise permitirá identificar áreas onde a presença de radares pode estar correlacionada com uma menor incidência de acidentes, fornecendo insights valiosos para políticas de segurança viária.


```python
import geopandas as gpd
from shapely.geometry import Point

# Convertendo os DataFrames para GeoDataFrames
gdf_radares = gpd.GeoDataFrame(df_radar01, geometry=gpd.points_from_xy(df_radar01.longitude, df_radar01.latitude))
gdf_acidentes = gpd.GeoDataFrame(df_prf01, geometry=gpd.points_from_xy(df_prf01.longitude, df_prf01.latitude))

# Calculando a distância mínima entre os acidentes e os radares mais próximos
def calcular_distancia_minima(x):
    return gdf_radares.distance(x).min()

# Criando a coluna 'distancia_radar_km' no DataFrame original
df_prf01['distancia_radar_km'] = gdf_acidentes.geometry.apply(calcular_distancia_minima)
```

```python
df_prf01['distancia_radar_km'].describe()
```

```python
# Definindo o número de bins (intervalos) no histograma
num_bins = 18

# Calculando os bins para o histograma
hist, bins = np.histogram(df_prf01['distancia_radar_km'], bins=num_bins)

# Calculando  a largura de cada bin
width = np.diff(bins)

# Calculando  o centro de cada bin
center = (bins[:-1] + bins[1:]) / 2

# Criando o histograma
fig, ax = plt.subplots(figsize=(12, 6))  # Ajuste as dimensões da figura
ax.bar(center, hist, align='center', width=width, color='#4D99FF', edgecolor='black')

# Adicionando rótulos e título ao gráfico
plt.xlabel('Distância até o radar mais próximo (km)', fontsize=12)
plt.ylabel('Número de Acidentes', fontsize=12)
plt.title('Distribuição da Ocorrência de Acidentes por Distância até o Radar', fontsize=14, fontweight='bold')

# Removendo as margens do gráfico
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Removendo a numeração do eixo y
plt.yticks([])

# Adicionando as quantidades de cada bin em cima das barras
for i in range(len(center)):
    if hist[i] > 0:
        ax.text(center[i], hist[i], str(int(hist[i])), ha='center', va='bottom', fontsize=10)

# Removendo as casas decimais dos valores no eixo X
plt.xticks(range(int(min(df_prf01['distancia_radar_km'])), int(max(df_prf01['distancia_radar_km']) + 1)), fontsize=10)

# Exibindo o histograma
plt.show()

```

Após a análise estatística dos dados, foram identifacado que:

-  A distância mínima entre um acidente e o radar mais próximo é de apenas 0,000025 km, enquanto a distância máxima é de 17,13 km. Os percentis indicam que 25% dos acidentes estão a uma distância de até 0,076 km do radar mais próximo, 50% estão a até 0,462 km e 75% estão a até 2,28 km.

- Ao examinar os dados do histograma, observa-se que a maioria dos acidentes (98.779) ocorre em uma faixa de distância entre 0 e 1 km do radar mais próximo. Em seguida, temos 16.624 acidentes que ocorreram em uma distância entre 1 e 2 km, seguidos por 11.375 acidentes na faixa de 2 a 3 km. Há uma queda gradual no número de acidentes à medida que a distância aumenta, com alguns picos notáveis em intervalos específicos. Por exemplo, observamos 3.313 acidentes na faixa de 3 a 4 km, 4.173 acidentes na faixa de 4 a 5 km e 3.817 acidentes na faixa de 6 km. Surpreendentemente, houve um aumento notável de acidentes na faixa de 7 km, totalizando 8.038 ocorrências.
